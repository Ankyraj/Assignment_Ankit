# Q1: Print all permutations of string.

```python
def fun(arr, l, r):
    if l==r:
        print(''.join(arr))
    else:
        for i in range(l,r+1):
            arr[l], arr[i] = arr[i], arr[l]
            fun(arr, l+1, r)
            arr[l], arr[i] = arr[i], arr[l]
            
            
strg = "Anky"
n = len(strg)
arr = list(strg)
fun(arr, 0, n-1)
 
```

# Q2: Implementation of LinkedList.

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, data):
       new_node = Node(data)
       if self.head is None:
            self.head = new_node
            return

       last = self.head
       while (last.next):
           last = last.next
       last.next =  new_node

        
    def printList(self):
        temp = self.head
        lst = []
        while temp != None:
            lst.append(temp.data)
            temp = temp.next
        return(lst)
        

llist = LinkedList()
llist.append(1)
llist.append(2);
llist.append(3);
llist.append(4)
print('Created linked list is:',llist.printList())
```
