# Question 1 – Replace Vowels with a digit

```python
#Sol:-
#Sum all the prime numbers between 1 and the resulting number i.e. 'n'
def summ_prime(n):
    # dp to store prime numbers
    dp = [True] * (n + 1)
    
    i = 2
    while i * i <= n:
        # If dp[i] is not changed, then
        # it is a prime
        if dp[i] == True:
            # Update all multiples of i
            j = i * 2
            while j <= n:
                dp[j] = False
                j += i
        i += 1

    # Return sum of primes generated through sieve.
    summ = 0
    for i in range (2, n + 1):
        if(dp[i]):
            summ += i
    return(summ)


# Add the digits of the number you get from the last step until you get a single digit
def digit_sum(n):
    summ = 0
    while(n > 0 or summ > 9):
     
        if(n == 0):
            n = summ
            summ = 0
         
        summ += n % 10
        n //= 10
     
    return summ

def fun(s):
    # ind store the index of vowel char that is present in the string
    ind = [i for i in range(len(s)) if s[i] in 'aeiou']
    
    # lst store multiply the index with 100
    mul = [i*100 for i in ind]
    
    # lst store the Sum all the prime numbers between 1 and the resulting number from mul list.
    lst = [summ_prime(i) for i in mul]
    
    # summ store Add the digits of the number you get from the last step until you get a single digit
    summ = [digit_sum(i) for i in lst]
    
    # new_s is the list in which I store the index value of vowel char and remaining char store as it is.
    new_s = [i if s[i] in 'aeiou' else s[i] for i in range(len(s))]
    
    # traverse the summ and ind list together with the help of zip fn to replace the vowel with the digit from the last step
    for i,j in zip(ind,summ):
        new_s[i] = j
        
    # return the output as a string with the join fn.
    return(''.join(map(str,new_s)))
```

```python
s1 = input() # first I/P
print(fun(s1))
```

```python
s2 = input() # second O/P
print(fun(s2))
```

# Question 2 – Minimum Swaps to Symmetric String

```python
#Sol:-

def swaps_cnt(s, n):
    s = list(s)
    # Counter to count minimum swap
    cnt = 0
    res = True
    
    # A loop which run in half string from starting
    for i in range(n // 2):
        l = i # left pointer
        r = n - l - 1 # right pointer

        while l < r:
            # if both char same then
            # break the loop if not
            # same then we have to move
            # right pointer to one step left
            
            if s[l] == s[r]:
                break
            else:
                r -= 1
                
        # it denotes both pointer same position and we don't have sufficient char to make palindrome string
        if l == r:
            res = False
            break
        else:
            for j in range(r, n - l - 1):
                (s[j], s[j + 1]) = (s[j + 1], s[j])
                cnt += 1
    if res:
        return (cnt)
    else:
        return(-1)
```

```python
s = input()
print(swaps_cnt(s, len(s)))
```
